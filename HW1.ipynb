```python
from PyQt5 import QtWidgets, QtGui, QtCore
from HW1_GUI import Ui_MainWindow
import sys
import numpy as np
import cv2
import numpy
```

```python
# 1.1 Load Image File 
def Load_Image():
    img1 = cv2.imread('Q1_Image/Uncle_Roger.jpg')
    h, w, c = img1.shape
    print('width:  ', w)
    print('height: ', h)
    cv2.imshow("Uncle Roger",img1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 1.2 Color Separation
def Color_Separation():
    img2 = cv2.imread('Q1_Image/Flower.jpg')
    (B,G,R) = cv2.split(img2)  
    zeros = np.zeros(img2.shape[:2], dtype = "uint8")
    cv2.imshow("Red",cv2.merge([zeros, zeros, R]))
    cv2.imshow("Green",cv2.merge([zeros, G, zeros]))
    cv2.imshow("Blue",cv2.merge([B, zeros, zeros]))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 1.3 Image Flipping  
def Image_Flipping():
    img1 = cv2.imread('Q1_Image/Uncle_Roger.jpg')
    flipped_img1 = cv2.flip(img1,1)
    cv2.imshow("Image Flipping",flipped_img1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 1.4 Blending
def Blending():
    def nothing(x):
        pass
    img1 = cv2.imread('Q1_Image/Uncle_Roger.jpg')
    flipped_img1 = cv2.flip(img1,1)
    img = np.zeros(img1.shape, dtype = "uint8")
    cv2.namedWindow('image')
    cv2.createTrackbar('Blend','image',0,255,nothing)
    while(1):
        cv2.imshow('image',img)
        k = cv2.waitKey(1) & 0xFF
        if k == 27:
            break
        
        r = cv2.getTrackbarPos('Blend','image')
        r = float(r)/255.0
        img = cv2.addWeighted(img1,r,flipped_img1,1.0-r,0)

    cv2.destroyAllWindows()
```

```python
# 2.1 Median filter
def Median_filter():
    img = cv2.imread('Q2_Image/Cat.png')
    img_medianBlur = cv2.medianBlur(img,7)
    cv2.imshow("Median image",img_medianBlur)
    cv2.imshow("Original image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#2.2 Gaussian Blur
def Gaussian_Blur():
    img = cv2.imread('Q2_Image/Cat.png')
    img_GaussianBlur = cv2.GaussianBlur(img,(3,3),0)
    cv2.imshow("Gaussian image",img_GaussianBlur)
    cv2.imshow("Original image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#2.3 Bilateral filter
def Bilateral_filter():
    img = cv2.imread('Q2_Image/Cat.png')
    img_bilateralFilter = cv2.bilateralFilter(img,9,90,90)
    cv2.imshow("Bilateral image",img_bilateralFilter)
    cv2.imshow("Original image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# 3.1 Gaussian Blur
def Gaussian_Blur_():
    img = cv2.imread('Q3_Image/Chihiro.jpg')
    img_gray = cv2.imread('Q3_Image/Chihiro.jpg',cv2.IMREAD_GRAYSCALE)
    img_gaussian = cv2.imread('Q3_Image/Chihiro.jpg',cv2.IMREAD_GRAYSCALE)
    outputimg = np.zeros([248, 459])
    gaussian_filter = np.array([[0.045,0.122,0.045],[0.122,0.332,0.122],[0.045,0.122,0.045]])

    for i in range(248):
        for j in range(459):
            outputimg[i,j] = gaussian_filter[0,0]*img_gaussian[i,j]+gaussian_filter[0,1]*img_gaussian[i,j+1]+gaussian_filter[0,2]*img_gaussian[i,j+2]+gaussian_filter[1,0]*img_gaussian[i+1,j]+gaussian_filter[1,1]*img_gaussian[i+1,j+1]+gaussian_filter[1,2]*img_gaussian[i+1,j+2]+gaussian_filter[2,0]*img_gaussian[i+2,j]+gaussian_filter[2,1]*img_gaussian[i+2,j+1]+gaussian_filter[2,2]*img_gaussian[i+2,j+2]

    outputimg = outputimg.astype(numpy.uint8)
    cv2.imshow("Original image",img)
    cv2.imshow("Grayscale image",img_gray)
    cv2.imshow("Gaussian image",outputimg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#3.2 Sobelx
def Sobelx():
    img = cv2.imread('Q3_Image/Chihiro.jpg')
    img_gaussian = cv2.imread('Q3_Image/Chihiro.jpg',cv2.IMREAD_GRAYSCALE)
    outputimg_gaussian = np.zeros([248, 459])
    gaussian_filter = np.array([[0.045,0.122,0.045],[0.122,0.332,0.122],[0.045,0.122,0.045]])

    for i in range(248):
        for j in range(459):
            outputimg_gaussian[i,j] = gaussian_filter[0,0]*img_gaussian[i,j]+gaussian_filter[0,1]*img_gaussian[i,j+1]+gaussian_filter[0,2]*img_gaussian[i,j+2]+gaussian_filter[1,0]*img_gaussian[i+1,j]+gaussian_filter[1,1]*img_gaussian[i+1,j+1]+gaussian_filter[1,2]*img_gaussian[i+1,j+2]+gaussian_filter[2,0]*img_gaussian[i+2,j]+gaussian_filter[2,1]*img_gaussian[i+2,j+1]+gaussian_filter[2,2]*img_gaussian[i+2,j+2]

    outputimg_gaussian = outputimg_gaussian.astype(numpy.uint8)
    cv2.imshow("Gaussian image",outputimg_gaussian)
    
    sobelx = outputimg_gaussian
    sobelx_filter = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    outputimg = np.zeros([246, 457])
    
    for i in range(246):
        for j in range(457):
            outputimg[i,j] = abs(sobelx[i,j]*sobelx_filter[0,0] + sobelx[i+1,j]*sobelx_filter[1,0] + sobelx[i+2,j]*sobelx_filter[2,0]
            + sobelx[i,j+2]*sobelx_filter[0,2] + sobelx[i+1,j+2]*sobelx_filter[1,2] + sobelx[i+2,j+2]*sobelx_filter[2,2])

    Max = np.amax(outputimg)
    Min = np.amin(outputimg)

    for i in range(246):
        for j in range(457):
            outputimg[i,j] = 255*(outputimg[i,j]-Min)/(Max - Min)

    outputimg = outputimg.astype(numpy.uint8)
    cv2.imshow("Original image",img)
    cv2.imshow("sobelx image",outputimg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#3.3 Sobely
def Sobely():
    img = cv2.imread('Q3_Image/Chihiro.jpg')
    img_gaussian = cv2.imread('Q3_Image/Chihiro.jpg',cv2.IMREAD_GRAYSCALE)
    outputimg_gaussian = np.zeros([248, 459])
    gaussian_filter = np.array([[0.045,0.122,0.045],[0.122,0.332,0.122],[0.045,0.122,0.045]])

    for i in range(248):
        for j in range(459):
            outputimg_gaussian[i,j] = gaussian_filter[0,0]*img_gaussian[i,j]+gaussian_filter[0,1]*img_gaussian[i,j+1]+gaussian_filter[0,2]*img_gaussian[i,j+2]+gaussian_filter[1,0]*img_gaussian[i+1,j]+gaussian_filter[1,1]*img_gaussian[i+1,j+1]+gaussian_filter[1,2]*img_gaussian[i+1,j+2]+gaussian_filter[2,0]*img_gaussian[i+2,j]+gaussian_filter[2,1]*img_gaussian[i+2,j+1]+gaussian_filter[2,2]*img_gaussian[i+2,j+2]

    outputimg_gaussian = outputimg_gaussian.astype(numpy.uint8)
    
    cv2.imshow("Gaussian image",outputimg_gaussian)
    
    sobely = outputimg_gaussian
    sobely_filter = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    outputimg = np.zeros([246, 457])
    
    for i in range(246):
        for j in range(457):
            outputimg[i,j] = abs(sobely[i,j]*sobely_filter[0,0] + sobely[i,j+1]*sobely_filter[0,1] + sobely[i,j+2]*sobely_filter[0,2]
            + sobely[i+2,j]*sobely_filter[2,0] + sobely[i+2,j+1]*sobely_filter[2,1] + sobely[i+2,j+2]*sobely_filter[2,2])

    Max = np.amax(outputimg)
    Min = np.amin(outputimg)

    for i in range(246):
        for j in range(457):
            outputimg[i,j] = 255*(outputimg[i,j]-Min)/(Max - Min)

    outputimg = outputimg.astype(numpy.uint8)
    cv2.imshow("Original image",img)
    cv2.imshow("sobely image",outputimg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 3.4 Magnitude
def Magnitude():
    img = cv2.imread('Q3_Image/Chihiro.jpg')
    img_gaussian = cv2.imread('Q3_Image/Chihiro.jpg',cv2.IMREAD_GRAYSCALE)
    outputimg_gaussian = np.zeros([248, 459])
    gaussian_filter = np.array([[0.045,0.122,0.045],[0.122,0.332,0.122],[0.045,0.122,0.045]])

    for i in range(248):
        for j in range(459):
            outputimg_gaussian[i,j] = gaussian_filter[0,0]*img_gaussian[i,j]+gaussian_filter[0,1]*img_gaussian[i,j+1]+gaussian_filter[0,2]*img_gaussian[i,j+2]+gaussian_filter[1,0]*img_gaussian[i+1,j]+gaussian_filter[1,1]*img_gaussian[i+1,j+1]+gaussian_filter[1,2]*img_gaussian[i+1,j+2]+gaussian_filter[2,0]*img_gaussian[i+2,j]+gaussian_filter[2,1]*img_gaussian[i+2,j+1]+gaussian_filter[2,2]*img_gaussian[i+2,j+2]

    outputimg_gaussian = outputimg_gaussian.astype(numpy.uint8)
    
    # x direction
    sobelx = outputimg_gaussian
    sobelx_filter = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])

    outputimg1 = np.zeros([246, 457])
    for i in range(246):
        for j in range(457):
            outputimg1[i,j] = abs(sobelx[i,j]*sobelx_filter[0,0] + sobelx[i+1,j]*sobelx_filter[1,0] + sobelx[i+2,j]*sobelx_filter[2,0]
            + sobelx[i,j+2]*sobelx_filter[0,2] + sobelx[i+1,j+2]*sobelx_filter[1,2] + sobelx[i+2,j+2]*sobelx_filter[2,2])


    Max = np.amax(outputimg1)
    Min = np.amin(outputimg1)

    for i in range(246):
        for j in range(457):
            outputimg1[i,j] = 255*(outputimg1[i,j]-Min)/(Max - Min)

    outputimg1 = outputimg1.astype(numpy.uint8)

    # y direction
    sobely = outputimg_gaussian
    sobely_filter = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

    outputimg2 = np.zeros([246, 457])
    for i in range(246):
        for j in range(457):
            outputimg2[i,j] = abs(sobely[i,j]*sobely_filter[0,0] + sobely[i,j+1]*sobely_filter[0,1] + sobely[i,j+2]*sobely_filter[0,2]
            + sobely[i+2,j]*sobely_filter[2,0] + sobely[i+2,j+1]*sobely_filter[2,1] + sobely[i+2,j+2]*sobely_filter[2,2])


    Max = np.amax(outputimg2)
    Min = np.amin(outputimg2)

    for i in range(246):
        for j in range(457):
            outputimg2[i,j] = 255*(outputimg2[i,j]-Min)/(Max - Min)

    outputimg2 = outputimg2.astype(numpy.uint8)

    # sum 
    outputimg = np.zeros([246, 457])
    for i in range(246):
        for j in range(457):
            outputimg[i,j] = (outputimg1[i,j]**2 + outputimg2[i,j]**2)**0.5
    outputimg = outputimg.astype(numpy.uint8)

    # show
    cv2.imshow("Sobelx image",outputimg1)
    cv2.imshow("Sobely image",outputimg2)
    cv2.imshow("sobel image",outputimg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def Transforms(Tx, Ty, Rotation, Scaling):
    img = cv2.imread('Q4_Image/Parrot.png')
    rows,cols = img.shape[:2]
    H = np.float32([[1,0,Tx-160],[0,1,Ty-84]])
    res1 = cv2.warpAffine(img,H,(cols,rows))
    M = cv2.getRotationMatrix2D((Tx,Ty),Rotation,Scaling)
    res = cv2.warpAffine(res1,M,(cols,rows))
    
    cv2.imshow("Transformation image",res)
    cv2.imshow("Original image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super(MainWindow, self).__init__()
        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)
        self.ui.PushButton.clicked.connect(Load_Image)
        self.ui.pushButton_5.clicked.connect(Color_Separation)
        self.ui.pushButton_6.clicked.connect(Image_Flipping)
        self.ui.pushButton_7.clicked.connect(Blending)
        self.ui.pushButton_3.clicked.connect(Median_filter)
        self.ui.pushButton_11.clicked.connect(Gaussian_Blur)
        self.ui.pushButton_13.clicked.connect(Bilateral_filter)
        self.ui.pushButton_2.clicked.connect(Gaussian_Blur_)
        self.ui.pushButton_8.clicked.connect(Sobelx)
        self.ui.pushButton_9.clicked.connect(Sobely)
        self.ui.pushButton_10.clicked.connect(Magnitude)
        self.ui.pushButton_12.clicked.connect(self.buttonClicked)
       
    def buttonClicked(self):
        Tx = int(self.ui.lineEdit_3.text())
        Ty = int(self.ui.lineEdit_4.text())
        Rotation = float(self.ui.lineEdit.text())
        Scaling = float(self.ui.lineEdit_2.text())
        Transforms(Tx, Ty, Rotation, Scaling)
        self.ui.lineEdit.clear()
        self.ui.lineEdit_2.clear()
        self.ui.lineEdit_3.clear()
        self.ui.lineEdit_4.clear()
        
```
